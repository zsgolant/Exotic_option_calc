The Monte Carlo numerical scheme is an approximation method widely used in options pricing. We use the Euler-Maruyama scheme for approximating numerical solutions of an SDE. The advantage of the method is that it enables us to approximate viable prices for complex derivatives, of higher order which do not have closed form solution with the Black-Scholes model. However, it is crucial to optimise the trade-off between the computing costs and practical feasibility of this pricing method. 

By simulating a path of stock prices according to the risk-neutral framework, we can caluclate the option's payoffs according to the Asian option specification. The option price is then the average of the discounted payoffs.

The report focuses on arithmetic average Asian options because the arithmetic average of lognormal distributions is not log normal and needs numerical approximation; unlike geometric average Asian options.

$V\ \left(S;\ t\right)=\ e^{-r\left(T-t\right)}E^Q\left[payoff\ (S_t)\right];\ r\ =\ risk\ free\ rate$


```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

Firstly, we define four option-pricing calculators. For each of them, we create a matrix where each column is a new simulation, and each row is a new day. Then, with the starting price=S0, we can evolve the process by the EM scheme. Then, for each simulation, we compute the payoff under this evolution of the stock price, discount it and take the average to get our Monte-Carlo estimate of the option price. The only difference in each of the following option-pricing calculators is the payoff structure, outlines below.

Below is the first of the four option-pricing calculators. It is for a fixed-strike Asian option. Fixed-strike options have a strike price determined in the contract, and the payoff is the difference between the average price and this strike. Firstly, we consider the continuous sampling (every day) average, such that the payoff of the option is:


$e^{-r\left(T-t\right)}E^Q\left[\max{\left(\frac{1}{T}\int_{0}^{T}S\left(\tau\right)d\tau-E,0\right)}\right]$


```python
def fixed_continuous(S0,E,T,sigma,r,dt,N):
    days=int(T/dt);
    rand_matrix=np.random.normal(size=(days,N));
    Stock_matrix=np.zeros((days,N));
    Stock_matrix[0,:]=S0;
    for row in range(1,len(Stock_matrix)):
        Stock_matrix[row,:]=Stock_matrix[row-1,:]*(1+r*dt+sigma*math.sqrt(dt)*rand_matrix[row-1,:]);
    option_payoff=np.zeros(N);
    for i in range(0,len(option_payoff)):
        option_payoff[i]=max(np.mean(Stock_matrix[:,i])-E,0);
    
    #The option price is calculated as discounted mean payoff, since under the risk-neutral measure, prices are martingales.
    option_price=np.mean(option_payoff)*math.exp(-r*T)
    return option_price
```

Next, we look to price a fixed-strike Asian option, under discrete sampling. The difference between this option and the previous one is that now, you input an array which marks the days you would like to sample from, and this is used to calculate the average from which the payoff is determined:

${e^{-r\left(T-t\right)}E}^Q[\max{\left(\bar{S}-E,0\right)}]$

```python
def fixed_discrete(S0,E,T,sigma,r,dt,N,sampling_dates):
    days=int(T/dt);
    rand_matrix=np.random.normal(size=(days,N));
    Stock_matrix=np.zeros((days,N));
    Stock_matrix[0,:]=S0;
    for row in range(1,len(Stock_matrix)):
        Stock_matrix[row,:]=Stock_matrix[row-1,:]*(1+r*dt+sigma*math.sqrt(dt)*rand_matrix[row-1,:]);
    option_payoff=np.zeros(N);
    
    #The difference with discrete sampling, we need to specify the dates from which to sample
    #in the function, include an array of days on which you want to sample
    #We need to account for 0 indexing system in Python
    sampling_indices=[i-1 for i in sampling_dates];
    for i in range(0,len(option_payoff)):
        option_payoff[i]=max(np.mean(Stock_matrix[sampling_indices,i])-E,0);
    option_price=np.mean(option_payoff)*math.exp(-r*T)
    return option_price
        
```

Next, we move onto pricing floating strike Asian options. The difference now is that no strike price is set out in the contract, it is determined at expiry as the average of the stock price over the lifetime of the option. Firstly, we look at continuous sampling average, such that the payoff of the option is:

$\ e^{-r\left(T-t\right)}E^Q\left[\max{\left(S-\frac{1}{T}\int_{0}^{T}S\left(\tau\right)d\tau,0\right)}\right]$

```python
def floating_continuous(S0,T,sigma,r,dt,N):
    days=int(T/dt);
    rand_matrix=np.random.normal(size=(days,N));
    Stock_matrix=np.zeros((days,N));
    Stock_matrix[0,:]=S0;
    for row in range(1,len(Stock_matrix)):
        Stock_matrix[row,:]=Stock_matrix[row-1,:]*(1+r*dt+sigma*math.sqrt(dt)*rand_matrix[row-1,:]);
    option_payoff=np.zeros(N);
    
    #Now the difference is to have the strike price as the mean of stock price
    for i in range(0,len(option_payoff)):
        option_payoff[i]=max(Stock_matrix[-1,i]-np.mean(Stock_matrix[:,i]),0);
    option_price=np.mean(option_payoff)*math.exp(-r*T)
    return option_price
```

And lastly, we set up the discrete-sampling equivalent for floating strike Asian options:

```python
def floating_discrete(S0,T,sigma,r,dt,N,sampling_dates):
    days=int(T/dt);
    rand_matrix=np.random.normal(size=(days,N));
    Stock_matrix=np.zeros((days,N));
    Stock_matrix[0,:]=S0;
    for row in range(1,len(Stock_matrix)):
        Stock_matrix[row,:]=Stock_matrix[row-1,:]*(1+r*dt+sigma*math.sqrt(dt)*rand_matrix[row-1,:]);
    
    #Again, now use discrete sampling
    option_payoff=np.zeros(N);
    sampling_indices=[i-1 for i in sampling_dates];
    for i in range(0,len(option_payoff)):
        option_payoff[i]=max(Stock_matrix[-1,i]-np.mean(Stock_matrix[sampling_indices,i]),0);
    option_price=np.mean(option_payoff)*math.exp(-r*T)
    return option_price
```

# Fixed vs Floating

```python
#Now that these have all been defined, we can compare and contrast 
# simply. First of all, lets compare the price of the floating
#and fixed strike option for continuous sampling
x=[i for i in range(0,200,10)]
y1=[floating_continuous(100,1,0.2,0.05,1/252,10000) ]
y2=[fixed_continuous(100,i,1,0.2,0.05,1/252,10000)for i in range(0,200,10)]
plt.axhline(y=y1, color='r', linestyle='--',label='Floating Strike')
plt.plot(x,y2, label='Fixed Strike')
plt.xlabel('Strike')
plt.ylabel('Price')
plt.show()
```

From the plot above, you can see that fixed strike options are more expensive when E<S0, less expensive when E>S0, and equally expensive for at-the-money option. By exploring the different payoffs of the option, we can explain this. Given that we have a constant, risk-neutral drift, we would expect the price at time T, to be the following:

$E^Q [S_T ]=S_0 e^{rT} $

Hence, we would expect the average stock price to be (given constant risk-neutral drift):

$S ̅=(S_0 (1+e^{rT} ))/2$

Now, we know that the payoff of the fixed strike Asian option to be:

$Payoff_{Fixed}=max⁡(S ̅-E,0)$

And the payoff of the floating strike Asian option to be:

$Payoff_{Floating}=max⁡(S_T-S ̅,0)$

Based on the definitions above, we can see that when $E=S_0$ we would expect that $S_T-S ̅=S ̅-E$ and so the prices of fixed and floating Asian options should be the same. When $E>S_0, S_T-S ̅>S ̅-E$ so the floating strike option should be more expensive, and when $E<S_0, S_T-S ̅<S ̅-E$ and the fixed strike option should be more expensive.



# Discrete vs Continuous

When we use discrete sampling in the pricing of the Asian option, the dates at which we sample will affect what we expect the average to be. If we take an early sample, the sampled average is likely to be lower than if we take a late average. Furthermore, if we take a representative average, we expect the sampled average to be similar to that of the continuously sampled average. Therefore, whether discrete or continuously sampled Asian options are more expensive will depend on the sampling dates. 

First of all, we consider a floating strike Asian option under three sampling methods. Early sampling (on days 1,5,10,15,20), representative sampling (on days 50,100,150,200,250) and late sampling (on days 230, 235, 240,245,250). Because we have a constant (positive) risk-neutral drift, we expect the averages in each of these methods to be in ascending order. Since the payoff of a floating price option is :

$max⁡(S_T-S ̅,0)$

This means that the payoff of option is decreasing in the size of the average, so the price under early sampling will generally be the largest, followed by representative sample and then late sample. This is demonstrated in the following table:


```python
x1=floating_discrete(100,1,0.2,0.05,1/252,10000,[1,5,10,15,20])
x2=floating_discrete(100,1,0.2,0.05,1/252,10000,[50,100,150,200,250])
x3=floating_discrete(100,1,0.2,0.05,1/252,10000,[230,235,240,245,250])
x4=floating_continuous(100,1,0.2,0.05,1/252,10000)
print(x1,x2,x3,x4)
```

| Early Sample | Representative Sample |  Late Sample | Continuous Sample |
| --- | --- | --- | --- |
| 10.11 | 5.03 | 1.53 | 5.72 |

Next, we move onto fixed strike Asian options. Under the same three discrete sampling methods, again the early sample will generally be the lowest, followed by the representative sample, and late sample will tend to be the highest. However, now the option payoff is:

$max⁡(S ̅-E,0)$

So, the option payoff is larger for larger sampling averages. This is demonstrated in the following plot, where the red line is the option price under early sampling, blue line is option price under representative sampling and black line is under late sampling. As you can see, for any strike price, the price of a fixed-strike Asian option is largest for late sample, then representative sample, then early sample.


```python

y2=[fixed_discrete(100,i,1,0.2,0.05,1/252,10000,[1,5,10,15,20])for i in range(0,200,10)]
y3=[fixed_discrete(100,i,1,0.2,0.05,1/252,10000,[50,100,150,200,250])for i in range(0,200,10)]
y4=[fixed_discrete(100,i,1,0.2,0.05,1/252,10000,[230,235,240,245,250])for i in range(0,200,10)]

fig, ax1 = plt.subplots(1, 1, figsize=(8, 10))



ax1.plot(x,y2,color='r')
ax1.plot(x,y3,color='b')
ax1.plot(x,y4,color='black')
ax1.set_title('Fixed Strike Sampling Comparison')

plt.show()
```

Notice in both fixed and floating Asian options, using a representative sample was effectively equivalent to the continuous sampling method. This demonstrates why discrete sampling Asian options are far more common, because you can create an option that is effectively equivalent to continuous sampling with much less computational effort.


# Number of Simulations

```python
#Next, We evaluate how the price changes as we change the number of
#simulations that we use. As you can see, as N increases, the price 
#starts to converge to some mean value.
x=[i for i in range(1000,100001,1000)];
y=[floating_continuous(100,1,0.2,0.05,1/252,i) for i in x];

plt.plot(x, y)

plt.xlabel('N')
plt.ylabel('Price')
plt.title('Floating Continuous')

plt.show()
```

```python
#And similarly for fixed continuous
x=[i for i in range(1000,100001,1000)];
y=[fixed_continuous(100,100,1,0.2,0.05,1/252,i) for i in x];

plt.plot(x, y)

plt.xlabel('N')
plt.ylabel('Price')
plt.title('Fixed Continuous')

plt.show()
```

The plots above show option price estimates as a function of N, for a Floating strike and Fixed strike Asian option respectively (both continuously sampled). It is important to note that the pricing calculator creates a new random sample for each N. This was done on purpose, to see how the estimates converge without reusing simulations. The number of simulations (N) is inversely proportional to the variance of the option price estimate, for both options.  

The reason that the estimates converge to the long term mean as we increase the number of simulations is because it is asymptotically consistent. 

Our estimate of the option price is: 

$\hat{P}=\frac{1}{N}\sum_{i}^{N}V\left(X_i\right)$

Where V(Xi) is the option value in the simulation denoted by Xi. Because prices are martingales under the risk neutral distribution:

$P=E^Q[V(X_i)]$

Using the Central Limit Theorem: 

$\hat{P}-P ~ N(0,\frac{\sigma_V^2}{N})$

Therefore, increasing the number of simulations increases the precision of our estimate. 

However, in practice, these accuracy gains and variability reduction have a lower bound limited by computational capacity. Some degree of variability in the option price estimate remains owing to the random number generation each time. In this specific option pricing model, the option price starts to converge to a stable value at around N = 60,000. Beyond this value of N, the tradeoff between marginal computational costs and marginal precision gains seem to nullify each other and thus best ignored in a practical context. 

The convergence rate may also vary depending on the complexity of the option pricing model. Nonetheless, with a suitably large N, the variability decreases, and the estimate becomes more precise. Overall, the graph demonstrates the importance of using a large number of simulations in the Monte Carlo method to obtain accurate and stable estimates of option prices. However, it is important to balance the computational costs and required precision depending on the practical objectives of the pricing model.  

# Changing parameter values

```python
#Finally, let's evaluate how the price changes with the different parameters
#vol
x1=np.arange(0, 1.1, 0.1);
y1=[floating_continuous(100,1,i,0.05,1/252,10000)for i in x1]

#IR
x2=np.arange(0, 0.25, 0.025);
y2=[floating_continuous(100,1,0.2,i,1/252,10000)for i in x2]

#T
x3=np.arange(0.5, 1, 0.1);
y3=[floating_continuous(100,i,0.2,0.05,1/252,10000)for i in x3]


fig, (ax1, ax2,ax3) = plt.subplots(3, 1, figsize=(8, 10))


ax1.plot(x1, y1)
ax1.set_title('Vol')

ax2.plot(x2, y2)
ax2.set_title('IR')

ax3.plot(x3, y3)
ax3.set_title('T')

plt.show()
```

Volatility  :

As you can see, increasing the underlying volatility increases the price of the option. The reason for this is that the option gives investor the right, but not obligation to buy the asset, and the probability that the option expires in the money increases as we increase volatility. Therefore, the likelihood that the option is exercised increases, and so it’s value increases.

Time to maturity : 

By a similar logic, increasing the time to maturity is going to increase the option price, as this also means the option is more likely to expire in the money. This is because under longer horizons (and a positive drift), the underlying is more likely to increase in price, and therefore expire in the money,

Interest rate :

Finally, let’s evaluate the effect of increasing interest rate. When r increases, the present value of the cash required to buy the call option, should it be exercised, is smaller, and therefore the present value of the option payoff is larger. You could equivalently think of this through the replicating portfolio. An option can be replicated by taking a leveraged long position in the underlying. When the interest rate is higher, you are required to borrow less at the risk-free rate, meaning that the portfolio is more expensive today. As such, the price of the option is higher.
